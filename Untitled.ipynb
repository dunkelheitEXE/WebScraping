## Importaciones

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
# url = "https://www.mercadolibre.com.mx/"
url = "https://listado.mercadolibre.com.mx/laptop#D%5BA:laptop%5D"
driver.get(url)
```

## Extracción de elementos

```python
li_elements = driver.find_elements(By.CSS_SELECTOR, '.shops__layout li')
data = []
```

## Guardamos elementos en una lista de datos

```python
for element in li_elements:
    try:
        element_title = element.find_element(By.TAG_NAME, 'h2')
        element_price = element.find_element(By.CSS_SELECTOR, '.andes-money-amount__fraction')
        element_image = element.find_element(By.CSS_SELECTOR, '.ui-search-result-image__element').get_attribute('src')
        data.append({
            'Title': element_title.text,
            'Price': f'${element_price.text}',
            'Image_reference': element_image
        })
    except NoSuchElementException:
        pass
```

## Verificamos que la informacion haya sido guardada satisfactoriamente

```python
df = pd.DataFrame(data)
df
```

## EXPORTAMOS LOS DATOS EN UN ARCHIVO CSV

```python
df.to_csv("Laptops.csv")
```

## QUITAMOS LA SESION DE DRIVER

```python
driver.quit()
```
